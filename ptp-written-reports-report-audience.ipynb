### Report Audience
*** 

But this is the most important part of the penetration test. 

It doesn't matter what you do during the penetration test, if you don't write it down, it wasn't worth doing. You need to make sure that your penetration testing reports make sense for the audience that it's targeted to. This does not just mean for the technical individuals within the organization. This needs to contain the structured details of what was done, how the organization can improve, and how the organization can use this for their next penetration test down the road. 

So, when you doing report writing, you need to make sure that you have an adequate amount of time set aside in order to produce the report. Now, some firms will have an actual technical writer that will take over the writing of the report and the penetration tester just needs to make sure that they pass along appropriate notes that they took during the engagement. However, in some other organizations, specifically smaller ones are organizations that are just starting to get into this space - you need to be able to write a report. 

Countless penetration testers are very good at the technical aspects of their job, but they cannot write a report to save their life that is able to be read by someone that is not a technical person, so, when it comes to doing your report, you need to ensure that you have enough time to do it properly and that it makes sense and that you have enough time to edit it after your first draft - don't just hand over the first draft. 

At the end of the day, this is probably 50% of the engagement time and 100% of what the organization is actually paying for. They're not specifically paying for you to poke holes in their network. They are more paying for you to show how they can improve their organization security posture.

When writing this, you need to consider that it's not going to just be consumed by your technical individual. Oftentimes, this is going to be read by individuals within the C-Suite, so, this means your executives within the organization, your CEO who's often in charge of the organization isn't going to have the time to read through a 200 page report but they are going to read the executive summary to make sure that they know what the individuals within their organization are doing.

The CFO is going to be looking at the finances of this - OK, that's great - we can improve in this area, but at what cost? So, you need to make sure that when you're writing it you have some idea in there, what the challenge of implementing a certain thing is along with the potential cost of it. 

Need to consider the Chief Operating Officer, the COO. You might propose something at the end of your penetration test in areas to improve, and the COO looks at it like no, that does not fit our operational model. But they are one of these individuals that you are writing the report for, so it needs to make sense to them - can't just be technical jargon. 

But if you do have technical jargon in there, that's what your Chief Information Officer is for. They are there to dive deep on the technical aspects with the specialist within the organization to figure out what solutions they can provide that will make them better as an organization. You need to make sure that you're including content specifically for every area of the business and every level of the business, so that everyone can consume this document, understand it and move on it. 

Every organization has key stakeholders that they interact with on a daily basis - these are typically individuals within the company that have a vested interest in the company or project. 

They don't necessarily have to be internal to the organization - can often be external. This could include something like investors, communities that interact with the company and suppliers that are providing the organization with things like materials or software or systems. 

The reason why you need to make sure that you understand who the organization's stakeholders there when you're writing your report is because oftentimes, they will share this with these third parties and these third parties need to understand it, or they will share the reason for implementing a project with these third parties and they reference the penetration testing report as the reason for it, so when you're doing this, you need to consider that some parts of your report may become public, may be shared with suppliers, may be shared with subcontractors or contractees - for example, if a company is providing services to a government, oftentimes, part of that contract is that they have to have a penetration test done and provide that report to the government. So with that, you need to make sure that it's being able to be understood by the government and the suppliers, investors and communities. 

Really have to make sure that this report makes sense to a layperson that may not necessarily be a technical individual. 

Do have to make sure that your technical staff can understand this, so, this is your IT management in your IT technical staff, your CIO, and individuals within the IT team, your networking team, your server-side team. They're going to have to understand what exactly they need to do to improve their systems. 

- How you got in? 


- What technique you used?


- How it was affective?


- Why it was effective?


- what patch needs to be applied?

These are all information that they're going to be looking for in the penetration test. 

Oftentimes that the penetration test report at the end when IT management goes through with the technical staff, they're going to say, look, there is five or six items on here, we can bang off in an afternoon. These are just simple patches to systems - can plug a few holes - this is going to be great - this penetration test was fantastic, will have to add this to our patch management strategy going forward. 

At the same time, there's going to be things in there that is going to require substantial changes possibly to the IT architecture or to IT practices internally and you need to make sure that the IT management, IT technical staff understand how things were conducted, so oftentimes, you're going to have several different sections to your report, depending upon how you're structuring it. 

Going to have the executive side, which is written in a way that a layperson can understand it and then the technical side - be afraid to get right down into the details with this stuff. This is what the IT technical team is looking for. It's even better if they can go and recreate the attack on their own and see how it happened.

Your developers are probably going to want to understand the tools that were used during the penetration testing exercises and how they worked. They're going to want to understand exactly how you got into their web applications that they built, or the applications that they built that are not necessarily hosted within the web. 

Now, the reason for this is because the developers oftentimes are going to be the ones that are going to have to integrate a solution, if it's a custom application or a custom solution that the organization has. 

Not every penetration test is going to be strictly network, email, social - now, you're also going to have internal applications your organization is built and has been using for years -some of them are legacy applications that have been around for decades that the organization is still using that summer student built back in 2001 - that's still working great for the organization - might be working great, but there might be significant security issues with it. 

So, this is what your development team is going to want to do. They're going to want to understand:

- What tools were used?


- How did they get into our applications? 


- How can we secure it going forward?
