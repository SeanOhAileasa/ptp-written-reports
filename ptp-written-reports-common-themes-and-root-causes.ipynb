### Common Themes and Root Causes
*** 

Now, one of the areas where you can sort of put all the information that you found, like any of the observations you had when during the penetration test is in the general findings section. 

Now, this is a synopsis of issues that have been discovered during the penetration test and this is really a general area, so you not going to get into specifics or into great detail in the general finding section, but you are going to be able to provide some information there that management can take and act upon. 

Really, the general findings area is targeted towards the manager as supposed to the technical section which is going to be targeted more to the technician, so, the manager can look to the general findings and say OK, we clearly our missing on our password policy - we're clearly missing on our social engagement or social awareness and we need to make sure we're doing a better job of ensuring that our firewalls are safe because there's a bunch of open ports here and then, they can just fire this off and the technical folks can take the technical part of the report and run with it.

Anytime you can use a graphical representation within your documentation - highly recommended - some of the things you might be able to get away with using a graphic for or when you're showing what targets were tested - specifically, if it's people within the organization and you were able to get their photo. 

Any processes that you tested, if you can find a good graphic that you can use to really hit home, what the process is and what was tested within the process - that's going to be very helpful to the person that's reading the report. 

Any attacks scenarios that were undertaken by you when you were doing the penetration test, so, for example, if you were hired to do a specific type of test against an organization, let's say a denial of service attack, well, you can make sure that then attacks scenario has some graphical representation. 

Of course any success rates that you had when doing the penetration test always look really good in a graph because you can show, especially, if this is the second or third penetration tests their organizations done - can show their improvement or if they're getting worse - for example, if you do a social engineering engagement with an organization and you notice that 80% of their staff, all fall for a spear phishing campaign ...well, that's going to look very good in a graphic but also, the organization clearly has some work to do.

Often very difficult to determine, but when you looking at the root cause of how you were able to get into a system, sometimes it's not as clear as you'd like it to be, so you're going to take a look and try to determine what the initial cause of the incident was. 

It may be that the system wasn't patched, it may be that the system had something in secure setup on it but occasionally, you will be a little bit lost because the root cause may be something within the config and you don't have access to that specific config because if you're doing a blind penetration test, oftentimes the root cause analysis is going to have to be conducted by someone on the other end of it - someone that's actually working for the organization. 

Anytime you do a root cause analysis, you need to make sure it's presented in an easy-to-read format. This is because management is going to look at the root cause to determine what course of action needs to be taken. Now, if it's a technical issue - great, you can fire that off to your technical people and will be able to implement it but if it's a policy or procedure change, it needs to be easy to understands so that there can be real action taken by the organization.

Oftentimes set up vulnerabilities as a completely different section within the penetration test, because while I may find lots of vulnerabilities on a corporate network, especially once I've gained access to it but not going to test them all, that's typically not part of the engagement. 

It is to report on where they can improve - now, they do want that done, you can certainly do it - can try to take every vulnerability and exploit it, but you're probably going to end up breaking their corporate network, so, some of the vulnerabilities you'll see are:


- Insecure configurations - this one's very common, especially if they're using older devices.


- Any command and code injections that can be done against a specific piece of software. 


- Patch management is probably the biggest one you'll run into because organizations may understand that they have to patch Windows every patch Tuesday but, they might not be doing their switches, their firewalls, their routers, their actual PC firmware, etc. 


- Any user awareness vulnerabilities that you find.


- Any unsupported software that's been installed within the organization by unsupported software - what I'm talking about here is software that is no longer supported by the organization that provides it, so it's no longer patched. It can't fall into that patch management strategy. You will see this from time to time, especially with legacy applications and sometimes it's a mission critical application.

Any observations you make during your penetration test, you're going to need to write down, and you're going to want to report on, so, some of the observations you'll make or your staff behavior - how people are reacting to you when you're onsite. 

If you're doing a social engineering engagement or how staff have behaved behind the scenes when you've conducted a test. So, let's say you ran an ``nmap`` scan of a network and you scanned everything very loudly and nothing happened - no one responded ...well, that's important to note because even if you weren't able to gain any access from that ``nmap`` scan that should be setting off alarm bells for any organization that has a significantly high security awareness level. 

What are the peak activity times for the organization, when is everybody the most busy because this is probably when they are the most vulnerable. At the other side of that coin, when are they least busy or when there's no one in the office. If no ones in the office after 5pm or Saturday and Sunday, the weekends probably the best time to do any type of attacking inside organization.

- What security controls were you able to detect that the organization has in place? 

So, if your ``nmap`` scan, tripped an intrusion detection system that's important to note or any security controls, you come acrossed on the machines themselves is going to be very helpful for the organization and for you because chances are if you're doing a penetration test on the organization today, you might be asked to come back and do one again a year later because some organizations will have these done at every year, or every quarter, or every half year.

Definitely want to make sure that you don't club organizations over the head with best practices because not all best practices are considered best practices by everybody but, there are some sort of universal ones in password management is probably the biggest one. 

Need to make sure that organizations are doing their password management properly - still telling organizations to this - see it all the time where someone high up in the organization has it set, so their password never expires and they like it that way, so, if that's the way it is, make sure they have multi-factor authentication turned on. 

Fine, you can keep that password forever, but here's a secure token that you're going to need to put into your computer in order to gain access and if you notice an organization as a poor password policy and no multi-factor authentication, you're going to want to include that in your report.

If there is a significant lack of policies within the organization, let's also a red flag. Now, some organizations will have lots of policies, but no one will have ever read them. They will simply gone through board approval, IT will signed off on and that is it, so, you want to make sure that when you're looking at this from a penetration testing standpoint when writing the report, that if there is any policies that you think the organization should have, make sure you mention it to them.
