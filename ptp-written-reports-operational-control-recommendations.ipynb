### Operational Control Recommendations
*** 

When writing your report, there's going to be some common operational control recommendations that you come back to time and time again. 

One of the big ones is job rotation - the idea behind job rotation is that every employee is cross-trained on something else, so there is a managerial approach that shifts employees between two or more assignments within the organization.

This is done to help reduce the monotony of the job and ensure that the individual gets larger experience within the company but, it is also done from a security standpoint to ensure that when someone goes out on vacation, someone else can fill in for them. 

One of the things you have to protect against (reminded an organization), they have to protect against is insiders to their organization conducting attacks and job rotation helps flush that out because when someone else takes on a role from someone else when they're out on vacation, that's when you'll often discover if they were doing anything that was typically incorrect or wrong. 

The job rotation is something that can be recommended from some organizations, but not all, so, keep in mind the size of the organization when you're making your recommendations from the penetration test.

When you are writing your recommendation, there's some objectives that you want when you are recommending job rotation. 

- Typically you want to expose employees to operations.


- Test employee competencies, so, if you have two people trained on one thing, suddenly you have a comparative there, and this allows you to test to see if the employees competent with that or not. 


- Going to develop more skill sets within your organization.


- Plan for succession - this is a big one and a lot of organizations really miss the boat on it - you need to make sure that you have a succession plan specifically when it comes to IT security. If an individual in a mission critical role gets hit by a bus tomorrow then What is the plan for succession? Have to have something there just to rescue the organization - in the real world, it's been that organizations often find a way, but the ones that have a real plan often do very well with this and a lot better in job rotation really allows you to plan for succession well. 


- Allows you to place employees within the organization better - this isn't just something from an IT security standpoint - this benefits the organization as a whole.

Another operational control that you'll often recommend is that the organization implement some type of time of day or day restrictions for log on. Now, what this does is this ensures that users can only access resources during certain hours of the day or certain days of the week and really what you're getting after here is you don't want traffic from here, users happening at 3 am while they're asleep. 

So, if you simply restrict it that helps improve your overall security, especially if it's a smaller organization and they can't monitor their security posture 24/7, so, they may not have monitoring after hours, so ensuring individuals are logged in only during the hours you are supposed to be that's going to help the organization as a whole created better security posture. 

That being said, this is a managerial practice that has to be initiated - you can't just implement this as an IT administrator, something that requires some level of communication to ensure that users know that this is coming into effect, so, a lot of users today will have significant bleed in between their personal and work lives, and that may mean that they are accessing their email late at night and actually responding to things and if you have users that are doing that, you want to make sure that they can still log in and do their work at those times of day. 

So, you want to make sure that if you are recommending this in a report that you remind them in the report of that managerial concern, so that you know the IT guy just isn't simply taking this report and just checking every box and banging them off one by one and then saying there, perfect, done and all of a sudden they have a bunch of users mad at them because you know they can't access any of their systems after 8pm.

Another operational control you can recommend is mandatory vacation. 

This is a policy that ensures that employees have to take time away from the job, so that someone else can step into that role for period of time to:

- Understand that role


- Make sure that things aren't being done nefariously in that role

So, this is a great way to pick up on financial fraud for example, so if you have an accountant that's embezzling funds and suddenly they go away for three weeks and you notice, our bottom lines a lot better than someone else is doing that job then you can investigate that. 

You know there is a lot of things you can do around mandatory vacation that's going to help the organization. Plus it also helps prevent burnout with employees, so, this is something that is very good for an organization to implement. There is no real cost to this because most organizations have vacation time, and it's going to improve the overall security posture of the organization.

Probably the recommendation that you will write the most about in every report you ever do is user education. 

User education is one of the key ways that organization can improve their overall security posture - this helps:

- Reduced errors


- Increases productivity


- Empowers employees to make the right decision


- Adds significant organizational value

User training is important but, it has to be done well, so, one of the things you often hear about when you go and do a penetration test, if you are doing user interviews as part of this after, you will often hear individuals saying, we received our IT security training and it was boring. This doesn't have to be boring - you can give individuals within the organization a taste of what it's like to live in the world of a penetration tester because this is not a boring field at all and those users will often see that as being something a little bit salacious that maybe, you know, seems a little bit wrong, and they'll pay attention. So, you want to make sure that you're doing your user training, and when you're are writing your report, this is something that will find its way into every report, because every organization can do this better.
