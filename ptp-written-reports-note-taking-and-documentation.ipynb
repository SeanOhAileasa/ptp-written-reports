### Note Taking and Documentation
*** 

Note taking is a very important topic, though it's not always covered off by penetration testing or when individuals teach penetration testing, so, one of the reasons why you want to take notes while you're doing a penetration test is so that you can learn to be a better penetration tester. 

When you're taking notes, you are creating connections in memory of what you did and this is going to help you because chances are something you've done on the penetration test today is going to be helpful again two, three, four times over the next year because oftentimes, you'll run into similar situations over and over again. 

When you write something down, it really helps you to recall it later, so, when you're going to write the report, it helps to know exactly what you did. 

If it's organized well, then you can share this with the group or the other penetration testers that were taking on this specific exercise. Now, when you share your notes what you're doing is, you're not only sharing specifics as it pertained to that penetration test, you're actually sharing knowledge with the group, so, a good note taking exercise will allow others in the group to search other notes even while they're doing a penetration test. For example, just came acrossed a server 2012 machine and notice that there is a piece of software that's running on it as a service - might be able to search through other penetration tester notes within the organization and find someone who's run into that situation before and it'll save me time because I'll know what steps to take in order to gain access to that machine.

Note taking really helps you improve your focus on the penetration test because you are always focused on what you're doing but you should be writing down what's working and what's not working as you're going through the exercise. 

It also improves your productivity - can go back and search through notes, and that's going to be helpful - that's going to make you more productive. The next time you run into that issue, you're going to be able to refer to your notes and oh look! this is the steps I took - take those steps again and maybe that will give me access to this machine.

Not only does note taking have to happen on paper, but a lot of individuals will take digital notes today. 

Now with digital notes, you have to make sure that it's stored well in a safe space because this is sensitive information on an organization, so, make sure your storage is encrypted and then it is possibly even put into a cloud-based system but ideally your notes should be searchable. 

Want to be able to search through those notes to potentially find things when not only you're writing your report, but when you're doing your penetration test as well. You can go back and look at things that have worked in the past. 

You're going to want to share your notes with others in the organization. Chances are, if you've done a penetration tester part of a team, you might not be the one writing the report, so, you're going to need to share your notes with someone else. 

Finally, you want to make sure that you're controlling access to these reports so that only the individuals have access to them - at the same time, you also want to make sure that they are removed after a certain period of time, so just as in the report that you give the organization is important, and ensuring that report does not become public is important. You don't want your notes becoming public either, so if you were able to get root on domain controller - well, OK great - but if I have your notes, I know how you did it now, so you want to make sure that access is very strictly controlled to these.

Some penetration testers use voice recognition software. They will carry a little voice recorder with them and take recordings as they're going. This is an alternate option for taking notes. This one typically is very good if you're doing any type of onsite penetration test because you may not be able to have the time when you're onsite to write things down, so having some type of voice recognition system where you can just make your verbal notes as you go, is going to help you out later when you go to write that report. 

There are some note taking applications out there - the key thing that you want to look for in a good note taking application is that it can do some form of formatting. 

You want this to be something that you can consume again later, so, you want it to be formatted well or in a standard that you can understand. 

You're going to want to be able to store images because you're going to have images that you're going to take of machines and of systems and of IPs and stuff and oftentimes, the print screen button is a lot quicker than writing down all the information on the screen. 

An audio recording, if you have an audio recording, attach it to the note taking application. 

Any attachments that you may have. So, if you have the version of the malware that worked, you may want to attach that into the note taking applications so that it can be provided to the organization at the end.

Any screenshots you may have taken need to be able to be put into that note taking application.

This is probably the most important thing about taking notes during the testing phase is that you should be taking your notes throughout the entire testing phase. 

Every time you attempt something, you should write down the time, what it was you attempted, and what the outcome was and the reason for this is because even if there wasn't a success, let's say, you didn't get root on that box, you weren't able to get a reverse shell but you were generating a lot of traffic on the network and you were not being very stealthy about it and no one seemed to respond. That's important too, because that allows the organization to improve. If they're not actively monitoring anything that might be something you'd want to note later in your report.

Oftentimes with notes, you're going to have multimedia and you need to make sure that you're treating this properly, so, you'll have screenshots sometimes - oftentimes if on a machine and have an output from an IP config or something, might just screenshot it because it's a lot easier than writing down all the information and then I'll go and write it down later, or when I go to update my notes and then, you'll have video clips as well. 

Some penetration testers will actually screen record while they're doing a penetration test - others when going onsite to do any type of social engineering engagement, they will actually carry a little video camera on them and a little audio recording and this really helps them go back and revisit what was successful what was not and it also allows these individuals to actually notice things that they may not have noticed when they're onsite. Like oftentimes, you're fairly aware of your surroundings, but if you're focused solely on gaining access to a specific area, you might not notice how employees are looking at you throughout the penetration test.
