### Securing Storage and Distribution of Reports
*** 

When you're writing your report, it's important to remember that the report needs to be handled securely and disposed of properly and the reason for this is because that penetration testing report is going to include some very sensitive information about the organization, systems, architecture, and ways to get access to them. 

So, your penetration test report outside of the executive summary is going to include a technical summary and this technical summary is going to have a ton of information here, about configs that were found, about systems that were penetrated, about what specific Metasploit modules were used, what vulnerabilities were found, what accounts were used for the penetration test. All of this information is great if you're an attacker to have because your work has already been done for you - know exactly the path to take to get access to the organization, so, you need to make sure that the report that is written is maintained in a secure fashion.

With handling the report, there are some best practices that you should ensure that you're using - for example, encrypt the digital copies - make sure when the penetration test is stored on a digital device that it is encrypted at rest. 

Any paper copies of the report and yes, there will be paper copies should be properly shredded and disposed of and tracked with the chain of custody when they're in use by your organization because keep in mind you're the penetration tester, not the one having the test done against you. How the organization handles the paperwork? Once you give it to them, that's up to them. Typically, they should follow these best practices as well but you should be sure that your organization is following these because you're responsible for ensuring that information stays safe from your perspective. 

When you are done with these reports, make sure you destroy the copies when they're no longer needed. This is because you want to make sure that you're not keeping these things around. If I can gain access to a penetration testing company, and I see a bunch of reports sitting on a shared drive, and I download them, I just gained access to probably 90% of those organizations as well, so, you need to make sure that these are properly destroyed or stored. Now, what some organizations will do is they will actually use secure offsite storage for these for long term storage, so, they will destroy the digital copy - will destroy the paper copies that exist within the organization but they will take a digital copy that is secured and they will take it to an offsite location, and stored in a secure vault where it is safe. So, if the organization does request another copy of it a couple years later, you can still go and get it and provide it, but you're not keeping it actively on your network.

Your penetration testing agreement is going to have a set amount of time you need to store that report for the organization conducted the penetration test against. The reason for putting this into writing is so that the organization knows the expectation of how long you will have a copy. 

Now, typically you will have your own records keeping governance that you'll require to by law, and you'll need to make sure you're following that as well but you don't want to keep this for example, on active storage onsite for five years. You might move it offsite after two weeks into non-active storage and in a secure vault where you can go, retrieve it if the organization ever requests it. But you want to make sure that it's agreed upon with the organization because one, they're going to want to make sure that this was actively handled properly, because it's a very sensitive document for that organization and two, you want to make sure that you're meeting the agreed standard from the organization but you also want to make sure that you're keeping your organization as safe as possible from possible litigation down the road. So if you get hit with a breach, well that's a problem but it's not as bigger problem if all of your penetration tests are held offsite where no one can gain access to them.

When you deliver these reports, you need to make sure, you're giving them to a proper stakeholder or executive within management. This is going to be detailed in the pre-engagement activities and agreements that you've undergone with the organization. So, that agreement is going to often say, who is required to have a copy of that report. Do not leave this on a secretary's desk. Do not leave this at the security front desk for the organization. Do not send this over email to someone in the organization. Make sure you're delivering this a hard copy and digital copy if you need to, through secure means only to the individuals that you've been contracted to supply to.

Storage of these reports is often going to be a requirement based upon your organizational's policy. Any compliance that the organization has to have for records keeping, and any regulations your organization has to have for records keeping. 

So, when you're storing this, you need to make sure that you're following the local laws that you have to follow for records keeping as an organization and you have to make sure you're compliant with any storage requirements that the organization is given you. 

You also need to make sure that you're following your own internal policies for proper document handling, so, these documents are incredibly sensitive - can't overstate just how sensitive these are. These are the instructions on how to break into that corporate network. That is not something you want to have just freely floating around the internet, so, you need to make sure that you're following the regulations and compliance. We have to follow to a T, and your organization's policies to a T and make sure that you're keeping that data as secure as possible because if that becomes public knowledge because someone within your penetration testing company accidentally emailed it externally or accidentally shared it on the cloud, that's a big problem.

One of the big benefits of cloud service providers is that they provide document storage, document management and document governance and lifecycle management as part of providing that cloud service. 

So, you can store these reports within a CSP. You just need to keep in mind that account breaches happen frequently within CSPs, so, you need to make sure that you are still auditing access and ensuring that you are monitoring access to those systems. That way you can ensure that, that data remains safe. 

Now typically once again, a best practice for this and maybe not everyone's is keep it for a week onsite somewhere encrypted - then move it offsite, take it away, put it on an encrypted thumb drive, put it in a safe somewhere like a safety deposit box in a bank under lock and key, and keep it there and that's probably the best way to store it because you can still go and get access to it if the customer requires it but at the same time, it's offsite - in a secure location and no one's going to be able to use an account to gain access. 

Now, someone might rob that bank and steal that thumb drive but they would still have to decrypt it in that case, sowith this stuff, it's important that you're following proper procedures and policies for your organization. If you're going to start within the cloud, you can just make sure that you're doing proper auditing and using lifecycle management on the document.

Once again, these are possibly some of the most sensitive documents the organization will come in contact with, so, you need to make sure that you're treating it with the utmost importance from a governance perspective.
